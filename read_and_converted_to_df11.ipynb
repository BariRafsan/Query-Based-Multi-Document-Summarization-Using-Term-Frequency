```python
#import required modules
import os
import pandas as pd
```

```python
data_set = r'D:\University\University 4,2\CSE475 Machine Learning\fwdduc20062007\DUC-2007\DUC-2007'

list_of_files = {} # dictionary
gold = {}
docsfile = {}

for (dirpath, dirnames, filenames) in os.walk(data_set): # oswalk itarate all posilbile paths#print(dirpath,dirnames, filenames)
    for filename in filenames:
        if not filename.endswith('.DS_Store'):
            file_p = os.sep.join([dirpath, filename]) # slashjoint
            topic = os.path.basename(os.path.normpath(dirpath))
            
            if not filename.endswith('.gold'):
                # docsfile['docsfile']
                # topic -> d30002t
                if docsfile.get(topic) == None:
                    docsfile[topic] = [ file_p ]
                else:
                    docsfile[topic].append(file_p)

            else:
                if gold.get(topic) == None:
                    gold[topic] = [ file_p ]
                else:
                    gold[topic].append(file_p)

list_of_files['docsfile'] = docsfile
list_of_files['gold'] = gold

list_of_files
```

```python
para = []
df_list = []
h_sum_list = []
g_1 = []
g_2 = []
g_3 = []
g_4 = []

for key, topic_dict in list_of_files.items():
    if key != 'gold':
        for topic_key, topic_value in topic_dict.items():
            s = ''
            for value_path in topic_value:
                with open(value_path, 'r') as f:
                    data = f.read()
                    s += data
                    s= s.replace('\n',' ')
                    s= s.replace('[','')
                    s= s.replace(']','')
                    
            df_list.append(topic_key)
            h_sum_list.append(key)
            para.append(s) 

    else:
        for topic_key, topic_value in topic_dict.items():
            with open(topic_value[0], 'r') as f:
                    data = f.read()
                    g_1.append(data)
                    with open (topic_key+'gold1.txt', 'w') as fp:
                                fp.write(str(g_1))
            with open(topic_value[1], 'r') as f:
                    data = f.read()
                    g_2.append(data)
            with open(topic_value[2], 'r') as f:
                    data = f.read()
                    g_3.append(data)
            with open(topic_value[3], 'r') as f:
                    data = f.read()
                    g_4.append(data)
for key, value in zip(df_list, para):
    #print(key, value)
    
    with open (key+'.txt', 'w') as fp:
                fp.write(str(value))
df2 = pd.DataFrame({ 'topic': df_list, 'paragraph': para, 'gold_1': g_1, 'gold_2': g_2, 'gold_3': g_3, 'gold_4': g_4 })
# print(df.iloc[0]['paragraph']) # first row and first column's value
df2.head(10) # first five
# df[(df['topic'] == 'd30001t') & (df['h_sum'] == 'gold')] # QUERY
```
